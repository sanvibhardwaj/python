```python
#declaration of list tuple dictionary
l=[]
k=[]
t=()
d={}
d1={}
d2={}

#function to add cars in list
def New_cars():
    a=int(input("Total Number of new deliveries:"))
    for i in range(a):
        s_no=int(input("Enter the serial number:"))
        brand=input("Enter the brand of car:")
        car_model=input("Enter the car model:")
        price=input("Enter the price:")
        print(30*"*")
        l.append([s_no,brand,car_model,price])
        k.append([brand,car_model])
    print(l)
    print(30*"*")


#function to show cars from the list
def show_cars():
    global t
    for i in l:
        print("The serial_no:",i[0])
        print("The brand:",i[1])
        print("The car_model:",i[2])
        print("The price:",i[3])
        print(20*"*")
        t=t+(i,)
    print(t)

#function to search a car in the list
def search_car():
    global d1
    b = int(input("Enter the serial number:"))
    for i in l:
        if i[0] == b:
            print("The car details are as follows:")
            d1[i[0]] = i[2]
            print(d1)
            return
    print("Sorry, The car is not available")

#function to check price
def check_price():
    global d2
    c = input("Enter the car model:")
    for i in l:
        if i[2] == c:
            d2[i[2]] = i[3]
            print(d2)
            return
    print("Sorry, The car is not available")

#function to remove car
def remove_car():
    global l
    d = input("Enter the car model:")
    l_copy = l[:]
    for i in l_copy:
        if i[2] == d:
            l.remove(i)
            print("The car is removed")
            return
    print("Car not found")

#function to calculate total amount to be paid by a buyer
def calculate_total_amount():
    tax_rate = 0.1
    total_amount = 0
    for car in l:
        price = float(car[3])
        total_amount += price + (price * tax_rate)
    print(f"The total amount to be paid for all cars in inventory including taxes is: ${total_amount:.2f}")

#for login to automobile
login_id=input("enter the ID")
password=input("enter the password")
if login_id == "car_store":
    if password == "1507":

#Main code
        while True:
            print("Continue to choosing a number:")
            print("1. To enter the car records")
            print("2. To show the car")
            print("3. To search for car models")
            print("4. To check the price for the car")
            print("5. To remove car models")
            print("6. To find total amounut")
            print("7. To Exit")
            ch = int(input("Enter the choice 1, 2, 3, 4, 5, 6,7: "))
            if ch == 1:
                New_cars()
            elif ch == 2:
                show_cars()
            elif ch == 3:
                search_car()
            elif ch == 4:
                check_price()
            elif ch == 5:
                remove_car()
            elif ch ==6:
                calculate_total_amount()
            else:
                break
    else:
        print("Incorrect password")
else:
    print("Incorrect login ID ")
```
